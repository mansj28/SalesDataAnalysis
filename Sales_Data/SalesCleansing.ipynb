### Sales Analysis

#### Import necessary libraries

```python
import pandas as pd
```

### Task 1. Read the sales data from year 2019 and merge files into one single file

```python
from glob import glob

sales_files = sorted(glob('Sales_*.csv'))
sales_files
```

```python
sales = pd.concat(pd.read_csv(file) for file in sales_files)
sales.head()
```

### Task 2. Cleansing the data. Check for null values

```python
print(sales.shape)
sales.isnull().sum()
```

```python
sales.dropna(inplace=True)
```

```python
sales.head()
```

```python
sales.shape
```

#### As we have merged our data we have the headers of each column added to the merged data

```python
sales[sales['Order ID'] == 'Order ID']
```

### this values are garbage values and needs to be removed

```python
index = sales[sales['Order ID'] == 'Order ID'].index
sales = sales.drop(index)
sales[sales['Order ID'] == 'Order ID']
sales.head()
```

### Check the column names. Change the names for easy accesibility

```python
sales.columns
```

```python
sales.columns = sales.columns.str.replace(' ','_')
```

```python
sales.head()
```

#### Check fot the dtypes and change where required

```python
sales.dtypes
```

#### We can convert order id, quantity, price to int and float. Date can be changed to datetime to perform certain analysis

```python
sales[['Order_ID', 'Quantity_Ordered']]  = sales[['Order_ID', 'Quantity_Ordered']].astype(int)
```

```python
sales['Price_Each'] = pd.to_numeric(sales['Price_Each'])
```

```python
sales['Order_Date'] = pd.to_datetime(sales['Order_Date'], format = '%m/%d/%y %H:%M') 
```

```python
sales.dtypes
```

### Task 3 Import the merged data to csv

```python
sales.to_csv('sales_data.csv',index = False)
```
